```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
X_train = np.load('z_train.npy')
X_test = np.load('z_test.npy')
X_train = np.squeeze(X_train)
X_test = np.squeeze(X_test)
y_train = np.load('train_label.npy', allow_pickle=True)
y_test = np.load('test_label.npy', allow_pickle=True)
```

```python
from autosklearn.experimental.askl2 import AutoSklearn2Classifier
import sklearn.metrics
```

```python
automl = AutoSklearn2Classifier(
    time_left_for_this_task=1800,
    per_run_time_limit=120,
)
automl.fit(X_train, y_train, dataset_name='imdb')
```

```python
print(automl.leaderboard())
```

```python
print(automl.show_models())
```

```python
predictions = automl.predict(X_test)
print("Accuracy score:", sklearn.metrics.accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
print('=======================================')
print(classification_report(y_test,predictions))
```

```python
predictions = automl.predict(X_train)
print("Accuracy score:", sklearn.metrics.accuracy_score(y_train, predictions))
```
